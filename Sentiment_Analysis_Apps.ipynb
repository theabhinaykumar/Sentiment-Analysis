# Sentiment Analysis on App Reviews
## Internship Project
This project analyzes user sentiment in app reviews using data visualization and statistical methods.

## 1. Data Loading & Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
apps_df = pd.read_csv('apps.csv')
reviews_df = pd.read_csv('user_reviews.csv')

# Display first few rows
reviews_df.head()
```

## 2. Handling Missing Values

```python
# Check for missing values
print(reviews_df.isnull().sum())

# Drop rows with missing sentiment values
reviews_df.dropna(subset=['Sentiment'], inplace=True)
```

## 3. Sentiment Distribution Analysis

```python
# Countplot of Sentiments
plt.figure(figsize=(6, 4))
sns.countplot(x='Sentiment', data=reviews_df, palette='coolwarm', hue='Sentiment', legend=False)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

## 4. Sentiment Polarity & Subjectivity Analysis

```python
# Scatter plot of Sentiment Polarity vs Subjectivity
plt.figure(figsize=(8,5))
sns.scatterplot(x=reviews_df['Sentiment_Polarity'], y=reviews_df['Sentiment_Subjectivity'], alpha=0.5)
plt.xlabel('Sentiment Polarity')
plt.ylabel('Sentiment Subjectivity')
plt.title('Sentiment Polarity vs Subjectivity')
plt.show()
```

## 5. Sentiment Trends Across Apps

```python
# Top 10 Apps with Most Reviews
top_apps = reviews_df['App'].value_counts().head(10).index
top_reviews = reviews_df[reviews_df['App'].isin(top_apps)]

# Sentiment distribution for top apps
plt.figure(figsize=(12,6))
sns.countplot(x='App', hue='Sentiment', data=top_reviews, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Sentiment Distribution for Top 10 Apps')
plt.show()
```
